# Cluster 6 — Model Usage Demo
Demonstrates that the C6 stacking model can be loaded and used in **one line of code**.

**Model:** LOO Stacking — LR + Decision Stump + KNN(k=1) → LR meta  
**Features:** 5  
**Train Accuracy:** 1.0 (5/5 bankruptcies caught)

```python
import pandas as pd
from cluster_6_classes import ClusterPredictor
```

```python
# Load cluster 6 training data (for demonstration)
df = pd.read_csv('../../Clusters/cluster_6.csv')
df.columns = df.columns.str.strip()
```

```python
# ── ONE LINE OF CODE ──────────────────────────────────────────
predictions = ClusterPredictor('c6_stacking_model.joblib').predict(df)
# ──────────────────────────────────────────────────────────────
```

```python
# Results
results = df[['Index', 'Bankrupt?']].copy()
results['Predicted'] = predictions
results['Correct']   = results['Bankrupt?'] == results['Predicted']
print(results.to_string(index=False))
print(f'\nCustom Accuracy: {(results[results["Bankrupt?"]==1]["Predicted"]==1).sum()} / {(results["Bankrupt?"]==1).sum()} bankruptcies caught')
```
