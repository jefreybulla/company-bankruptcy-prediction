# Cluster 0 — Model Usage Demo
Demonstrates that the C0 stacking model can be loaded and used in **one line of code**.

**Model:** SMOTE + Stacking — LR + RF + GBM + SVC → LR meta  
**Features:** 10  
**Threshold:** 0.49 (safe for test sparsity constraint)

```python
import pandas as pd
from cluster_0_classes import ClusterPredictor
```

```python
# Load cluster 0 training data (for demonstration)
df = pd.read_csv('../../Clusters/cluster_0.csv')
df.columns = df.columns.str.strip()
```

```python
# ── ONE LINE OF CODE ──────────────────────────────────────────
predictions = ClusterPredictor('c0_stacking_model.joblib').predict(df)
# ──────────────────────────────────────────────────────────────
```

```python
# Results
results = df[['Index', 'Bankrupt?']].copy()
results['Predicted'] = predictions
print(f'Total predicted bankrupt: {predictions.sum()} / {len(predictions)} ({predictions.mean()*100:.1f}%)')
print(f'Sparsity safe: {predictions.mean()*100:.1f}% < 20% limit')
bankrupt_df = results[results['Bankrupt?']==1]
print(f'\nBankrupt companies:')
print(bankrupt_df[['Index','Bankrupt?','Predicted']].to_string(index=False))
```
