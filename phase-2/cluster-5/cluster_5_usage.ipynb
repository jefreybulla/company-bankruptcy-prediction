# Cluster 5 Usage:

This notebook demonstrates how to load and use the exported predictor for cluster 5.

The predictor is a dict-based bundle containing:
- selected_features: 5 financial features picked via L1 logistic regression
- stacking_model: fitted StackingClassifier (LR + RF + GB base models, LR meta-learner)
- threshold: tuned decision threshold
- n_features: feature count for Table 3 reporting

## Setup

```python
import pandas as pd
import joblib

# Import helper functions
from aania_cluster_classes import predict_with_bundle, predict_proba_with_bundle

# Load the bundle
c5_bundle = joblib.load('aania_cluster5_model.joblib')

print('Cluster 5 predictor loaded')
print(f"  Features:  {c5_bundle['n_features']}")
print(f"  Threshold: {c5_bundle['threshold']:.2f}")
```

## Selected features

```python
for f in c5_bundle['selected_features']:
    print(f'  - {f.strip()}')
```

## Load Data

```python
df = pd.read_csv('../../Clusters/cluster_5.csv')
X = df.drop(columns=['Bankrupt?'])
y = df['Bankrupt?']
```

## Make Predictions

```python
y_pred = predict_with_bundle(c5_bundle, X)

print(f'Predictions made for {len(y_pred)} samples')
print(f'\nPrediction distribution:')
print(f'  Non-Bankrupt (0): {(y_pred == 0).sum()}')
print(f'  Bankrupt (1): {(y_pred == 1).sum()}')
```

## Sanity check against training notebook

Should match the Table 3 numbers from Aania_C5.ipynb (TT=66, TF=6).

```python
TT = ((y == 1) & (y_pred == 1)).sum()
TF = ((y == 1) & (y_pred == 0)).sum()
print(f'TT={TT}, TF={TF}, N_features={c5_bundle["n_features"]}')
```
